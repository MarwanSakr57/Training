```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
df = pd.read_csv('Weather Dataset.csv')
df.drop_duplicates(inplace=True)
df.dropna(how='all')
```

```python
df=df.sort_values(by ='Formatted Date',ascending =True)
df = df.dropna()
columns= [ "Humidity", "Wind Speed (km/h)", "Wind Bearing (degrees)", "Visibility (km)", "Pressure (millibars)"]
for column in columns:
    df = df[df[column] >= 0]
df=df[df['Temperature (C)']>-50]
df=df[df['Apparent Temperature (C)']>-50]
```

```python
x = df['Formatted Date']
y = df ['Apparent Temperature (C)']
plt.figure(figsize=(10, 5))
plt.plot(x,y)
plt.title('Temperature (C)')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.show()
```

```python
plt.hist(df['Apparent Temperature (C)'],bins =10,color='Purple',edgecolor='Black')
plt.title('Distribution of Temperature ')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['Humidity'],df['Apparent Temperature (C)'])
plt.title("Temperature Vs Humidity")
plt.ylabel('Temperature')
plt.xlabel('Humidity')
```

```python
col= [ "Temperature (C)","Apparent Temperature (C)","Humidity", "Wind Speed (km/h)", "Wind Bearing (degrees)", "Visibility (km)", "Pressure (millibars)"]
data = df[col]
```

```python
plt.figure(figsize=(10, 8))
sn.heatmap(data.corr(),annot=True ,cmap ='hot')
plt.title('Heatmap of Weather Dataset')
plt.show()
```
